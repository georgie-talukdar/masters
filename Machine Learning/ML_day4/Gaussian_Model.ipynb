```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from sklearn import mixture
%matplotlib inline  

n_samples = 30
print('sample number:', n_samples)
np.random.seed(0)


X_train = np.random.randn(n_samples, 2) + np.array([20, 20])
```

```python
print('two visualisation functions')
def Vis_points(_X):
    plt.xlim(15, 25)
    plt.ylim(15, 25)
    plt.scatter(_X[:, 0], _X[:, 1], .8)
    plt.show()
    
def Vis_Gaussian(_X):    
    x = np.linspace(15, 25)
    y = np.linspace(15, 25)
    X, Y = np.meshgrid(x, y)
    XX = np.array([X.ravel(), Y.ravel()]).T
    Z = -clf.score_samples(XX)
    Z = Z.reshape(X.shape)

    CS = plt.contour(X, Y, Z, norm=LogNorm(vmin=1.0, vmax=1000.0),
                 levels=np.logspace(0, 3, 10))
    CB = plt.colorbar(CS, shrink=0.8, extend='both')
    plt.scatter(_X[:, 0], _X[:, 1], .8)

    plt.title('Gaussian Distribution')
    plt.axis('tight')
    plt.show()
```

```python
print('visualising the points')
Vis_points(X_train)
```

```python
print('train the Gaussian model, and visualise the points')
clf = mixture.GaussianMixture(n_components=1, covariance_type='full')
clf.fit(X_train)
Vis_Gaussian(X_train)
```

```python
print('using the trained Gaussian model to generate new data samples')
n_generated = 10000
print('number of generated samples:', n_generated)
X_generated = clf.sample(n_generated)[0]
```

```python
print('visualising the generated data points')
Vis_Gaussian(X_generated)
```
